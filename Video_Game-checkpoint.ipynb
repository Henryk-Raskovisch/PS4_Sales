```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("PS4_GamesSales.csv", encoding = "unicode-escape")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize = (14, 5))
plt.title("Check Null Fields")
sns.heatmap(df.isnull(), cbar = False);
```

```python
#removing null values
df.dropna(inplace = True)
```

```python
df.describe()
```

```python
plt.title("Number of Global Sales (mi)", loc = "center", fontsize = "14")

sns.barplot(data = df, x = "Year", y = "Global", ci = None, color = "#69b3a2", estimator = sum)

plt.ylabel("Number of Sales (mi)")
```

```python
#withdraw the years 2019 and 2020
df = df.loc[(df["Year"] != 2019) & (df["Year"] != 2020)]

df.head()
```

```python
plt.figure(figsize = (12, 5))

plt.style.use("ggplot")

plt.title("Global Sales Distribution",
            loc = "left", fontsize = 14)

sns.kdeplot(df["Global"],
            shade = True, bw = 1,
            color = "#96a8a8",
            linewidth = 2.5);
```

```python
analyse = df.groupby(by = ["Year"]).sum().reset_index()
analyse.head()
```

```python
plt.figure(figsize = (10, 6))

plt.title("Global Distribution Analysis (mi)")

sns.boxplot(data = df, x = "Year", y = "Global");
```

```python
df.loc[df["Global"] >= 10]
```

```python
#analysing the proportion of 100% from each continent compared to the total
North_America = [america / total * 100 for america, total in zip(analyse["North America"], analyse["Global"])]
Europe = [europe / total * 100 for europe, total in zip(analyse["Europe"], analyse["Global"])]
Japan = [japan / total * 100 for japan, total in zip(analyse["Japan"], analyse["Global"])]
Rest_of_World = [world / total * 100 for world, total in zip(analyse["Rest of World"], analyse["Global"])]

North_America, Europe, Japan, Rest_of_World
```

```python
#size
plt.figure(figsize = (10, 5))

#bar width in graph
bar_width = 0.85
labels = analyse["Year"]
groups = [0, 1, 2, 3, 4, 5]

#title
plt.title("Analysing Distribution by Continent")

#america plot
plt.bar(groups, North_America, width = bar_width, color = "#b5ffb9", edgecolor = "white")

#Europe plot
plt.bar(groups, Europe, bottom = North_America, width = bar_width, color = "#f9bc86", edgecolor ="white")

#japan plot
plt.bar(groups, Japan, bottom = [a + b for a, b in zip(North_America, Europe)], width = bar_width, color = "#a3acfe", edgecolor = "white")

#rest of world plpot
plt.bar(groups, Rest_of_World, bottom = [a + b + c for a, b, c in zip(North_America, Europe, Japan)], width = bar_width, color = "#d3acf3", edgecolor = "white")

#labels
plt.xticks(groups, labels)
plt.xlabel("Group")
plt.ylabel("Distribution %")

#legend
plt.legend(["North America", "Europe", "Japan", "World"], loc = "upper left", bbox_to_anchor = (0.15, -0.1), ncol = 4 );
```

```python
from sklearn.preprocessing import LabelEncoder
label_function = LabelEncoder()

df["Number_Publisher"] = label_function.fit_transform(df["Publisher"])
df["Number_Genre"] = label_function.fit_transform(df["Genre"])
df["Number_Game"] = label_function.fit_transform(df["Game"])
df.head()
```

```python
palette_color = sns.color_palette("husl", 8)
palette_color
```

```python
#pd.concat([df["Number_Publisher"],df["Publisher"]]).unique()
a = df["Number_Publisher"].unique()
b = df["Publisher"].unique()
for i, j in zip(a, b):
    print(i, j)
```

```python
plt.figure(figsize = (15,30))
plot = sns.barplot(data = df, x = "Number_Publisher", y = "Publisher")

for i in plot.patches:
    plot.annotate(i.get_width(),
                  (i.get_width(), i.get_y() + i.get_height() / 2),
                  ha = "center", va = "baseline", fontsize = 12,
                  color = "black", xytext = (15, -5),
                  textcoords = "offset points")
```

```python
plt.figure(figsize = (20,5))
plt.title("Analysing Game Publisher (mi)")
sns.scatterplot(data = df, x = "Number_Publisher", y = "Global", color = palette_color[0]);
```

```python
plt.figure(figsize = (20,5))
plt.title("Analysing Game Genres (mi)")
sns.scatterplot(data = df, x = "Number_Genre", y = "Global", color = palette_color[0]);
```

```python
plt.figure(figsize = (20,5))
plt.title("Analysing Games (mi)")
sns.scatterplot(data = df, x = "Number_Game", y = "Global", color = palette_color[0]);
```
